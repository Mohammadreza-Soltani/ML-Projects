```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pca.fit(iris.data[:,[0,2]])
pca_transformed=pca.transform(iris.data[:,[0,2]])

mean=pca.mean_
first_pc=pca.components_[0]
plt.scatter(iris.data[:,0],iris.data[:,2])
plt.arrow(mean[0],mean[1],first_pc[0],first_pc[1],color="red",width=0.1)
plt.show()
```

```python
plt.scatter(pca_transformed[:,0],pca_transformed[:,1])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=Axes3D(fig)
versicalr=iris.data[50:100]
versicalr=iris.data[:,[0,1,2]]

ax.scatter(iris.data[:,0],iris.data[:,1],iris.data[:,2],c="r")
ax.set_title("iris versicalr")
ax.set_xlabel("sepal length")
ax.set_ylabel("sepal width")
ax.set_zlabel("petal length")   
plt.show()
```

```python
vc_pca=PCA()
vc_pca.fit(versicalr)
vc_transformed=vc_pca.transform(versicalr)

nfeaturs=range(vc_pca.n_components_)
var_features=vc_pca.explained_variance_

plt.bar(nfeaturs,var_features)
plt.xlabel("nfeaturs")
plt.ylabel("var_features")
plt.xticks(nfeaturs)
plt.show()
```

```python
dim_r=PCA(n_components=2)
dim_r.fit(iris.data)
dim_r_transformed=dim_r.transform(iris.data)

print(dim_r_transformed)
```

```python
lable=iris.target
plt.scatter(dim_r_transformed[:,0],dim_r_transformed[:,1],c=lable)
plt.show()
```
